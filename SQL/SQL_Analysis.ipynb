Import Libraries

```python
import pandas as pd
import sqlite3
```

Load the Cleaned Dataset

```python
df = pd.read_csv("SuperMarket_Analysis_Cleaned.csv")
```

Create an SQLite Database

```python
# Create SQLite database in memory
conn = sqlite3.connect(":memory:")
```

Convert the DataFrame into an SQL Table

```python
df.to_sql(
    "SuperMarket_Data",
    conn,
    if_exists="replace",
    index=False
)
```

View the Table

```python
query = """
SELECT *
FROM SuperMarket_Data
LIMIT 5;
"""

pd.read_sql(query, conn)
```

Revenue by Month (Line Chart Data)

```python
query = """
SELECT
    strftime('%Y-%m', Date) AS Month,
    ROUND(SUM(Revenue), 2) AS Total_Revenue
FROM SuperMarket_Data
GROUP BY Month
ORDER BY Month;
"""

monthly_revenue = pd.read_sql(query, conn)

monthly_revenue
```

```python
query = """
SELECT *
FROM SuperMarket_Data
LIMIT 1;
"""

pd.read_sql(query, conn)
```

Total Expenses by Product Line

```python
query = """
SELECT
    Product_Line,
    ROUND(SUM(Expense), 2) AS Total_Expense
FROM SuperMarket_Data
GROUP BY Product_Line
ORDER BY Total_Expense DESC;
"""

expense_product = pd.read_sql(query, conn)

expense_product
```

Total Profit by City

```python
query = """
SELECT
    City,
    ROUND(SUM(Profit), 2) AS Total_Profit
FROM SuperMarket_Data
GROUP BY City
ORDER BY Total_Profit DESC;
"""

profit_city = pd.read_sql(query, conn)

profit_city
```

Revenue by Branch

```python
query = """
SELECT
    Branch,
    ROUND(SUM(Revenue),2) AS Total_Revenue
FROM SuperMarket_Data
GROUP BY Branch
ORDER BY Total_Revenue DESC;
"""

branch_revenue = pd.read_sql(query, conn)

branch_revenue
```

Revenue by Payment Method

```python
query = """
SELECT
    Payment,
    ROUND(SUM(Revenue),2) AS Total_Revenue
FROM SuperMarket_Data
GROUP BY Payment
ORDER BY Total_Revenue DESC;
"""

payment_revenue = pd.read_sql(query, conn)

payment_revenue
```

Revenue by Product Line

```python
query = """
SELECT
    Product_Line,
    ROUND(SUM(Revenue),2) AS Total_Revenue
FROM SuperMarket_Data
GROUP BY Product_Line
ORDER BY Total_Revenue DESC;
"""

product_revenue = pd.read_sql(query, conn)

product_revenue
```

Revenue by Customer Type

```python
query = """
SELECT
    Customer_Type,
    ROUND(SUM(Revenue),2) AS Total_Revenue
FROM SuperMarket_Data
GROUP BY Customer_Type;
"""

customer_revenue = pd.read_sql(query, conn)

customer_revenue
```

Average Customer Rating

```python
query = """
SELECT
    Product_Line,
    ROUND(AVG(Rating),2) AS Average_Rating
FROM SuperMarket_Data
GROUP BY Product_Line
ORDER BY Average_Rating DESC;
"""

rating = pd.read_sql(query, conn)

rating
```

Overall Profit Margin

```python
query = """
SELECT
    ROUND(
        (SUM(Profit) * 100.0) / SUM(Revenue),
        2
    ) AS Profit_Margin_Percentage
FROM SuperMarket_Data;
"""

profit_margin = pd.read_sql(query, conn)

profit_margin
```

Executive KPI Query

```python
query = """
SELECT

COUNT(*) AS Total_Transactions,

ROUND(SUM(Revenue),2) AS Total_Revenue,

ROUND(SUM(Expense),2) AS Total_Expense,

ROUND(SUM(Profit),2) AS Total_Profit,

ROUND(AVG(Revenue),2) AS Average_Order_Value,

ROUND(AVG(Rating),2) AS Average_Rating

FROM SuperMarket_Data;
"""

kpis = pd.read_sql(query, conn)

kpis
```
