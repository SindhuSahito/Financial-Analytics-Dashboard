Import Required Libraries

```python
# Import Pandas
import pandas as pd
```

Load the CSV File

```python
# Load the dataset
df = pd.read_csv("SuperMarket Analysis.csv")

# Display the first 5 rows
df.head()
```

```python
# Rename columns for better readability
df.rename(columns={
    "Sales": "Revenue",
    "cogs": "Expense",
    "gross income": "Profit",
    "Customer type": "Customer_Type",
    "Product line": "Product_Line",
    "Unit price": "Unit_Price",
    "Invoice ID": "Transaction_ID",
    "Tax 5%": "Tax"
}, inplace=True)

# Display updated column names
print(df.columns.tolist())
```

Check the Dataset Structure

```python
# Display dataset information
df.info()
```

```python
# Display the number of rows and columns
print("Dataset Shape:", df.shape)
```

```python
# Display data types
df.dtypes
```

Display Missing Values

```python
# Check missing values
df.isnull().sum()
```

Remove Duplicate Transaction IDs

```python
# Count duplicate Invoice IDs
print("Duplicate Invoice IDs:", df["Invoice ID"].duplicated().sum())
```

```python
# Remove duplicate Invoice IDs
df = df.drop_duplicates(subset="Invoice ID")

# Verify duplicates are removed
print("Duplicate Invoice IDs After Cleaning:", df["Invoice ID"].duplicated().sum())
```

Handle Missing Revenue and Expense Values

```python
# Fill missing values in Sales (Revenue)
df["Sales"] = df["Sales"].fillna(df["Sales"].median())

# Fill missing values in cogs (Expense)
df["cogs"] = df["cogs"].fillna(df["cogs"].median())

# Verify missing values
df[["Sales", "cogs"]].isnull().sum()
```

```python
print(df.columns.tolist())
```

Remove Currency Symbols and Convert to Numeric

```python
# Remove currency symbols and commas from Sales
df["Sales"] = (
    df["Sales"]
    .astype(str)
    .str.replace(r"[$₹,]", "", regex=True)
)

# Convert Sales to numeric
df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")
```

```python
# Remove currency symbols and commas from cogs
df["cogs"] = (
    df["cogs"]
    .astype(str)
    .str.replace(r"[$₹,]", "", regex=True)
)

# Convert cogs to numeric
df["cogs"] = pd.to_numeric(df["cogs"], errors="coerce")
```

```python
print(df[["Sales", "cogs"]].dtypes)
```

Convert the Date Column

```python
# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])
```

```python
print(df["Date"].dtype)
```
